# MCTS Data Collection

This notebook runs the MCTS collection script to gather game data for RL training.

```python
from scoundrel.rl.mcts.collect import collect_games
from scoundrel.rl.mcts.evaluate_collection import evaluate_collection
from pathlib import Path
```

```python
# Configuration
num_games = None  # None = run until interrupted
seed = None  # None = random seed
num_simulations = 100000  # MCTS simulations per move
verbose = True
```

```python
# Run collection
results = collect_games(
    num_games=num_games,
    seed=seed,
    num_simulations=num_simulations,
    output_dir=None,  # Uses default: mcts/logs/collected_games
    verbose=verbose,
)
```

```python
# Display results using evaluate_collection
evaluate_collection(
    collection_dir=None,  # Uses default: mcts/logs/collected_games
    verbose=False,
)
```
